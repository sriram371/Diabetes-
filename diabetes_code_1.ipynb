```python
import pandas as pd

diabetes_df=pd.read_csv('diabetes.csv')
```

```python
y=diabetes_df['Outcome']
y
diabetes_df.drop(columns='Outcome',inplace=True)
diabetes_df # validation dataset
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(diabetes_df,y,test_size=0.2,random_state=57)

X_train
```

```python
from sklearn.linear_model import LogisticRegression
logreg=LogisticRegression()
logreg.fit(X_train,y_train)
predictions=logreg.predict(X_test)
predictions
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, recall_score,precision_score,f1_score

confusion_matrix(y_test,predictions)

accuracy_score(y_test,predictions)

recall_score(y_test,predictions)

precision_score(y_test,predictions)

f1_score(y_test,predictions)
```

```python
n_estimates=list(range(20,200,20))
```

```python
from sklearn.ensemble import RandomForestClassifier

accuracy=[]
for val in n_estimates:
    rndm_frt_cfr=RandomForestClassifier(n_estimators=val)
    rndm_frt_cfr.fit(X_train,y_train)
    accuracy.append(rndm_frt_cfr.score(X_train,y_train))
accuracy_rndm_frst=accuracy

```

```python
from matplotlib.pyplot import plot
plot(n_estimates,accuracy)
```

```python
from sklearn.ensemble import AdaBoostClassifier

accuracy_ada=[]
for val in n_estimates:
    ada_cfr=AdaBoostClassifier(n_estimators=val)
    ada_cfr.fit(X_train,y_train)
    accuracy_ada.append(ada_cfr.score(X_train,y_train))
    
accuracy_ada
```

```python
plot(n_estimates,accuracy_rndm_frst,'red')
plot(n_estimates,accuracy_ada,'blue')
```

```python
from sklearn.cluster import KMeans
from matplotlib import pyplot as plt 
kmean=KMeans()
kmean.fit(X_train)

# finding the optimal K value or optimal number of clusters
num_clusters=range(2,11)

wss=[]

for n in num_clusters:
    kmeans=KMeans(n_clusters=n)
    kmeans.fit(X_train)
    wss.append(kmeans.inertia_)
    
plt.plot(num_clusters,wss)
plt.title('Elbow Curve')
    
```

```python
k_means_1=KMeans(n_clusters=7)
k_means_1.fit(X_train)

k_means_1.predict(X_test)
```
